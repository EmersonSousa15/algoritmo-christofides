```python
from odd_vertices import odd_vertices
from prim import prim 
from hierholzer import hierholzer
import ast
import time
```

```python
import networkx as nx
import matplotlib.pyplot as plt


def generateNxGraph(edges):
    G = nx.Graph()
    
    for u, v, w in edges:
        G.add_edge(u, v, weight=w)
        
    return G
        
def plot_graph(G):
        
    # Desenha o grafo
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1500, font_size=16)
    nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): d['weight'] for u, v, d in G.edges(data=True)})
    
    plt.show()
    
```

1. Leitura do arquivo de entrada

```python
with open("entrada.txt", "r", encoding="utf-8") as arquivo:
    entrada = arquivo.readlines()
    
n = int(entrada[0].strip())
graph = [ast.literal_eval(line.strip()) for line in entrada[1:]]

if(len(graph[0]) > n):
    n = len(graph[0])

print("Matriz de adjacência recebida:", graph)
```

```python
# Tranformando em uma lista de arestas

E = []
V = []

for u in range(n):
    V.append(u)
    for v in range(n):
        peso = graph[u][v]
        E.append((u, v, peso))

G = generateNxGraph(E)

print(f"G = (V,E)")
print(f"V = {V}")
print(f"E (u,v,w) = {E}")

#plot_graph(G)
```

1. Geração da árvore geradora mínima T de G

```python
start = time.time()
T = prim(graph, n)
end = time.time()

print("Tempo de execução do algoritmo de Prim:", end - start,"segundos")

print("MST (u,v,w) =", [(u, v, {'weight': w}) for u, v, w in T])

s = 0
for t in T:
    s += t[2]

mstNx = generateNxGraph(T)

print("Peso da árvore geradora mínima: ", s)
#plot_graph(mstNx)
```

2. Identificação do conjunto 𝐼 de vértices de grau ímpar em 𝑇 (sabe-se que tal conjunto deve possuir uma quantidade par de elementos); 

```python
I = odd_vertices(T)
print("Vértices ímpares:",I)
```

3. Construção de um emparelhamento 𝑀 (também chamado de acoplamento ou casamento) 
perfeito de custo mínimo no sugrafo induzido pelos vértices de 𝐼;

```python
sub_graph = G.subgraph(I).copy()

start = time.time()
matching = nx.algorithms.matching.min_weight_matching(sub_graph, weight='weight')
end = time.time()

print("Tempo de execução do algoritmo de emparelhamento:", end - start, "segundos")

custo_total = sum(graph[u][v] for u, v in matching)

print("Vértices escolhidos:", I)
print("Emparelhamento:", matching)
print("Custo total do emparelhamento:", custo_total)

G_matching = [(u, v, graph[u][v]) for u, v in matching]

plot_graph(generateNxGraph(G_matching))
```

4. União de 𝑇 e 𝑀, combinando suas arestas, formando um multigrafo 𝐻 em que cada vértice possui grau par; 

```python
#Lista de adjacência do multigrafo H, com um dicionárioa no qual chave será um vertice e o valor uma lista de vizinhos desse vértice
H = {}

#Adciono as arestas de T tanto na lista de u como de v, dado que é um grafo não direcionado
#setdefault é um método de dicionário que serva para inicializar uma chave com um valor padrão se ela ainda não existir
for u, v, _ in T:
    H.setdefault(u, []).append(v)
    H.setdefault(v, []).append(u)

#Adiciona as arestas de matching, da mesma forma
for u, v in matching:
    H.setdefault(u, []).append(v)
    H.setdefault(v, []).append(u)

print("Multigrafo H =", H)
```

```python
start = time.time()
eulerian_cicle = hierholzer(H)
end = time.time()
print("Tempo de execução do algoritmo de Hierholzer:", end - start, "segundos")

print("Ciclo euleriano:", eulerian_cicle)
```

```python
#Cria uma lista para armazenar o ciclo hamitoniano
hamitoniano_cicle = []

#Função para remover os vértices repetidos do ciclo euleriano
#set() cria uma estrutura de dados que não armazena valores repetidos
#percorro o eulerian_cicle e verifico se já encontrei aquele vétice anteriormente, senão adiciono no hamitoniano_cicle e no encountared
def remove_vertices():
    encountared = set()
    for v in eulerian_cicle:
            if v not in encountared:
                hamitoniano_cicle.append(v)
                encountared.add(v)
                
remove_vertices()
print("O ciclo que visita cada vértice exatamente um vez, com o custo mínimo possível:", hamitoniano_cicle)
```
